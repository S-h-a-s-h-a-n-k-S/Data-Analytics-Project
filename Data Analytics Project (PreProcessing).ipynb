## Importing Pandas and reading the dataset

```python
import pandas as pd
df = pd.read_csv('Top_IT_companies_in_India.csv')
df.head(1)
```

### Using shape attribute to know the no. of rows and columns in the dataset.

```python
df.shape
```

### Using dropna() function to remove rows containing Null values if any.

```python
df.dropna(inplace=True)
```

### Using drop() function to drop the 'about' column.

```python
df.drop(columns=['about'], inplace=True)
```

```python
df.head(1)
```

```python
df.shape
```

### Using idxmax() function to find the company with the maximum rating.

```python
max_rating_company = df.loc[df['rating'].idxmax()]
```

```python
max_rating_company
```

### Filtering out the companies with rating >= 3.5

```python
df = df[df['rating'] >= 3.5]
```

```python
df.shape
```

### Filtering out the companies with founded_year >= 1968

```python
df = df[df['founded_year'] >= 1968]
```

```python
df.shape
```

### Filtering out the companies with total employees >= 1000

```python
df = df[df['total_emp'] >= 1000]
```

```python
df.shape
```

# EDA ANALYSIS

### Importing necessary libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Summary Statistics 

```python
df.describe()
```

### Data Visualization 

### --> Histogram to visualize the distribution of numerical variables.

```python
df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

### --> Bar Plot to compare categorical variables.

```python
sns.countplot(x='company_type', data=df)
plt.xticks(rotation=45)
plt.show()
```

### --> Scatter Plot to explore relationships between numerical variables.

```python
sns.scatterplot(x='rating', y='total_emp', data=df)
plt.show()
```

### --> Box Plot to identify outliers in numerical data.

```python
sns.boxplot(x='company_type', y='rating', data=df)
plt.xticks(rotation=45)
plt.show()
```

### Correlation Analysis

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```
